1. Find the highest rated movie in the “Quest” story type.

2. Find the genre in which there has been the greatest number of movie releases

4. Is there any correspondence between the critics’ evaluation of a movie and its
acceptance by the public? Find out, by plotting the net profitability of a movie
against the ratings it receives on Rotten Tomatoes.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/HollywoodMovies.csv')
data['NetProfitability'] = data['WorldGross'] / data['Budget']

# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(data['RottenTomatoes'], data['NetProfitability'], color='blue', alpha=0.7)

# Add labels and title
plt.title("Net Profitability vs Rotten Tomatoes Ratings", fontsize=14)
plt.xlabel("Rotten Tomatoes Rating (%)", fontsize=12)
plt.ylabel("Net Profitability (World Gross / Budget)", fontsize=12)

# Annotate points with movie names
for i, row in data.iterrows():
    plt.annotate(row['Movie'], (row['RottenTomatoes'], row['NetProfitability']), fontsize=9, alpha=0.7)

plt.grid(alpha=0.4)
plt.show()
```

  1. From the raw data below create a Pandas Series
'Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat'
a) Print all elements in lower case
b) Print all the elements in upper case
c) Print the length of all the elements
  

```python
import pandas as pd
import numpy as np

data = ['Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat']
series = pd.Series(data)

lower_case_series = series.str.lower()
print("All elements in lower case:")
print(lower_case_series)

upper_case_series = series.str.upper()
print("\nAll elements in upper case:")
print(upper_case_series)

length_series = series.str.len()
print("\nLength of all elements:")
print(length_series)
```

2. From the raw data below create a Pandas Series
' Atul', 'John ', ' jack ', 'Sam'
a) Print all elements after stripping spaces from the left and right
b) Print all the elements after removing spaces from the left only
c) Print all the elements after removing spaces from the right only



```python
import pandas as pd

data = [' Atul', 'John ', ' jack ', 'Sam']
series = pd.Series(data)

stripped_series = series.str.strip()
print("Elements after stripping spaces from both sides:")
print(stripped_series)

left_stripped_series = series.str.lstrip()
print("\nElements after removing spaces from the left only:")
print(left_stripped_series)

right_stripped_series = series.str.rstrip()
print("\nElements after removing spaces from the right only:")
print(right_stripped_series)
```

  6. Create a series and reverse all lower case words
'india 1998', 'big country', np.nan


```python
import pandas as pd
import numpy as np

def reverse_lowercase_words(text):
    if pd.isna(text):
        return text
    words = text.split()
    reversed_words = [word[::-1] if word.islower() else word for word in words]
    return ' '.join(reversed_words)

data = ['india 1998', 'big country', np.nan]
series = pd.Series(data)
reversed_series = series.apply(reverse_lowercase_words)
print(reversed_series)
```

  9. Create pandas series and print in three columns value 0 or 1 is a or b or c exists
in values
'a', 'a|b', np.nan, 'a|c'


```python
import pandas as pd
import numpy as np

data = ['a', 'a|b', np.nan, 'a|c']
series = pd.Series(data)
result = pd.DataFrame({
    'contains_a': series.str.contains('a', case=False, na=False).astype(int),
    'contains_b': series.str.contains('b', case=False, na=False).astype(int),
    'contains_c': series.str.contains('c', case=False, na=False).astype(int)
})
print(result)
```

  10. Create pandas dataframe having keys and ltable and rtable as below -
'key': ['One', 'Two'], 'ltable': [1, 2]
'key': ['One', 'Two'], 'rtable': [4, 5]
Merge both the tables based of key

```python
import pandas as pd

ltable = pd.DataFrame({
    'key': ['One', 'Two'],
    'ltable': [1, 2]
})
rtable = pd.DataFrame({
    'key': ['One', 'Two'],
    'rtable': [4, 5]
})
merged_df = pd.merge(ltable, rtable, on='key')
print(merged_df)
```

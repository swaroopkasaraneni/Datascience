1. You are given a dataset, which is present in the LMS, containing the number of
hurricanes occurring in the United States along the coast of the Atlantic. Load the
data from the dataset into your program and plot a Bar Graph of the data, taking
the Year as the x-axis and the number of hurricanes occurring as the Y-axis.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Hurricanes.csv')
print(df.head())

plt.bar(df["Year"], df["Hurricanes"], color='skyblue')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes per Year')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2. The dataset given, records data of city temperatures over the years 2014 and 2015. Plot the histogram of the temperatures over this period for the cities of
San Francisco and Moscow.


```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/CityTemps.csv')
print(df.head())

sf_temps = df[(df['Year'] == 2014) | (df['Year'] == 2016)]['San Francisco']

moscow_temps = df[(df['Year'] == 2014) | (df['Year'] == 2016)]['Moscow']

plt.figure(figsize=(12, 6))

plt.hist(sf_temps, bins=20, alpha=0.5, label='San Francisco', color='blue')
plt.hist(moscow_temps, bins=20, alpha=0.5, label='Moscow', color='orange')

plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Histogram of Temperatures in San Francisco and Moscow (2014-2015)')
plt.legend()
plt.show()
```

3. Plot a pie-chart of the number of models released by every manufacturer,
recorded in the data provide. Also mention the name of the manufacture with
the largest releases.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Cars2015.csv')
print(df.head())

car_type_counts = df['Type'].value_counts()
top_car_type = car_type_counts.idxmax()
top_count = car_type_counts.max()

print(f"Most frequent car type: {top_car_type} ({top_count} cars)")

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    car_type_counts,
    labels=car_type_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Distribution of Car Types')
plt.show()
```

4. Create csv file from the data below and read in pandas data frame
  1. Phase 1 -Reading Data
  2. Phase 2 –Describe the data
Describe the data on the unit price
  3. Phase 3 –filter the data
Create new dataframe having columns 'name','net_price','date' and group all the
records according to name
  4. Phase 4 –Plotting graph
Plot the graph after calculating total sales by each customer. Customer name should be
on x axis and total sales in y axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/dataVisual.csv')
print(df.head())

# Describe the 'unit price' column
print("Description of Unit Price:")
print(df['unit price'].describe())

# Select relevant columns
filtered_df = df[['name', 'net_price', 'date']]

# Group by 'name' and calculate total sales
grouped_data = filtered_df.groupby('name')['net_price'].sum().reset_index()
grouped_data.rename(columns={'net_price': 'total_sales'}, inplace=True)

# Display the grouped data
print(grouped_data)


# Plot the graph
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['name'], grouped_data['total_sales'], color='lightblue')

# Add labels and title
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.title('Total Sales by Customer')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

5. Let the x axis data points and y axis data points are
X = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

  1. Draw a Simple plot
  2. Configure the line and markers in simple plot
  3. configure the axes
  4. Give title of Graph & labels of x axis and y axis
  5. Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]
  6. define width, height as figsize=(4,5) DPI and adjust plot dpi=100
  7. Give a font size of 14


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data for the plot
X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]
y_error = [0.12, 0.13, 0.2, 0.1]

# Configure plot dimensions and DPI
plt.figure(figsize=(4, 5), dpi=100)

# Plot with line, markers, and error bars
plt.errorbar(X, y, yerr=y_error, fmt='-o', color='blue', label='Line with error bars', markersize=8, capsize=5)

# Configure the axes
plt.xlim(0, 5)
plt.ylim(19.5, 22)

# Add labels, title, and legend
plt.xlabel("X-axis", fontsize=14)
plt.ylabel("Y-axis", fontsize=14)
plt.title("Simple Plot with Error Bars", fontsize=14)
plt.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

 8. Draw a scatter graph of any 50 random values of x and y axis


```python
# Generate 50 random values for x and y
np.random.seed(42)
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 20

# Scatter plot
plt.figure(figsize=(6, 5), dpi=100)
plt.scatter(x_scatter, y_scatter, color='green', alpha=0.7, edgecolor='k')

# Add labels and title
plt.xlabel("Random X values", fontsize=14)
plt.ylabel("Random Y values", fontsize=14)
plt.title("Random Scatter Plot", fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

9. Create a dataframe from following data
'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]
Draw a Scatterplot of preTestScore and postTestScore, with the size of each point
determined by age
  

```python
# Create the DataFrame
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)

# Scatterplot with size by age
plt.figure(figsize=(6, 5), dpi=100)
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'] * 10, alpha=0.7, color='blue', edgecolor='k')

# Add labels and title
plt.xlabel("Pre-Test Score", fontsize=14)
plt.ylabel("Post-Test Score", fontsize=14)
plt.title("Scatterplot: Size by Age", fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

10. Draw a Scatterplot from the data in question 9 of preTestScore and
postTestScore with the size = 300 and the color determined by sex

```python
# Scatterplot with size=300 and color by gender
plt.figure(figsize=(6, 5), dpi=100)

# Map gender to color
colors = df['female'].map({0: 'blue', 1: 'red'})

plt.scatter(df['preTestScore'], df['postTestScore'], s=300, c=colors, alpha=0.7, edgecolor='k')

# Add labels and title
plt.xlabel("Pre-Test Score", fontsize=14)
plt.ylabel("Post-Test Score", fontsize=14)
plt.title("Scatterplot: Color by Gender", fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

1. Plot Total Sales Per Month for Year 2011. How the total sales have increased
over months in Year 2011. Which month has lowest Sales?



```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/content/BigMartSalesData.csv'
data = pd.read_csv(file_path)

data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%d-%m-%y')

data_2011 = data[data['InvoiceDate'].dt.year == 2011]
monthly_sales = data_2011.groupby(data_2011['InvoiceDate'].dt.month)['Amount'].sum()

# Plot the data
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales Per Month for 2011', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Identify the month with the lowest sales
lowest_month = monthly_sales.idxmin()
lowest_sales = monthly_sales.min()

print(f"The month with the lowest sales in 2011 is {lowest_month} with total sales of {lowest_sales}.")
```

2. Plot Total Sales Per Month for Year 2011 as Bar Chart. Is Bar Chart Better to
visualize than Simple Plot?


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/BigMartSalesData.csv'
data = pd.read_csv(file_path)
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%d-%m-%y')
data_2011 = data[data['InvoiceDate'].dt.year == 2011]
monthly_sales = data_2011.groupby(data_2011['InvoiceDate'].dt.month)['Amount'].sum()

# Plotting Bar Chart
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales Per Month for 2011 (Bar Chart)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plotting Line Plot
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Total Sales Per Month for 2011 (Line Plot)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(True)
plt.show()

lowest_month = monthly_sales.idxmin()
lowest_sales = monthly_sales.min()

print(f"The month with the lowest sales in 2011 is {lowest_month} with total sales of {lowest_sales}.")
```

3. Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest
towards sales?


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/BigMartSalesData.csv'
data = pd.read_csv(file_path)
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%d-%m-%y')
data_2011 = data[data['InvoiceDate'].dt.year == 2011]
country_sales = data_2011.groupby('Country')['Amount'].sum()


plt.figure(figsize=(8, 8))
country_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3', legend=False)
plt.title('Sales Contribution by Country in 2011', fontsize=14)
plt.ylabel('')
plt.show()


highest_country = country_sales.idxmax()
highest_sales = country_sales.max()

print(f"The country with the highest sales contribution in 2011 is {highest_country} with total sales of {highest_sales}.")
```

4. Plot Scatter Plot for the invoice amounts and see the concentration of amount.
In which range most of the invoice amounts are concentrated

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/BigMartSalesData.csv'
data = pd.read_csv(file_path)
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%d-%m-%y')
data_2011 = data[data['InvoiceDate'].dt.year == 2011]

# Plotting Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data_2011.index, data_2011['Amount'], alpha=0.5, color='green')
plt.title('Scatter Plot of Invoice Amounts in 2011', fontsize=14)
plt.xlabel('Invoice Index', fontsize=12)
plt.ylabel('Invoice Amount', fontsize=12)
plt.grid(True)
plt.show()



amount_range = data_2011['Amount'].describe()
print(amount_range)
```

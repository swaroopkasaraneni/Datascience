1.Display first 100 records from the data and play with parameters -tiles, zoom_start etc. Map will be saved in BasicWebMap.html â€“view it in browser


```python
import folium
import pandas as pd


data = pd.read_csv('/content/Police_Department_Incidents_Year_2016_.csv')


first_100 = data.head(100)
map_center = [first_100['X'].mean(), first_100['Y'].mean()]
m = folium.Map(location=map_center, zoom_start=12)

folium.TileLayer('OpenStreetMap').add_to(m)

for index, row in first_100.iterrows():
    folium.Marker(
        location=[row['X'], row['Y']],
        popup=row['Category']
    ).add_to(m)

m.save("BasicWebMap.html")

print("Map has been saved as BasicWebMap.html.")
```

2.For latest 7 days of data create WebMap of Crimes which are categorized as ROBBERY

```python
import folium
import pandas as pd


data = pd.read_csv('/content/Police_Department_Incidents_Year_2016_.csv')

data['Date'] = pd.to_datetime(data['Date'])
unique_dates = data['Date'].dt.date.unique()
sorted_unique_dates = sorted(unique_dates)
last_7_unique_dates = sorted_unique_dates[-7:]
print(last_7_unique_dates)

robbery_data = data[(data['Category'] == 'ROBBERY') & (data['Date'].dt.date.isin(last_7_unique_dates))]

map_center = [robbery_data['X'].mean(), robbery_data['Y'].mean()]
m = folium.Map(location=map_center, zoom_start=12)

# Add markers for each robbery
for _, row in robbery_data.iterrows():
    folium.Marker(
        location=[row['X'], row['Y']],
        popup=f"Date: {row['Date']}<br>Category: {row['Category']}",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(m)

m
```

3.For latest 15 days of data create One WebMap of Crimes which are categorized as FRAUD and GAMBLING. Change the icon to font awesome icon (http://fontawesome.io/icons/)

```python
import folium
import pandas as pd


data = pd.read_csv('/content/Police_Department_Incidents_Year_2016_.csv')

data['Date'] = pd.to_datetime(data['Date'])
unique_dates = data['Date'].dt.date.unique()
sorted_unique_dates = sorted(unique_dates)
last_15_unique_dates = sorted_unique_dates[-15:]
print(last_15_unique_dates)

filtered_data = data[(data['Category'].isin(['FRAUD', 'GAMBLING'])) & (data['Date'].dt.date.isin(last_15_unique_dates))]

map_center = [filtered_data['X'].mean(), filtered_data['Y'].mean()]
m = folium.Map(location=map_center, zoom_start=12,tiles="OpenStreetMap")

icon_fraud = 'money-bill'
icon_gambling = 'cogs'

# Add markers for each robbery
for _, row in filtered_data.iterrows():
    icon = icon_fraud if row['Category'] == 'FRAUD' else icon_gambling
    color = 'red' if row['Category'] == 'FRAUD' else 'blue'
    folium.Marker(
        location=[row['X'], row['Y']],
        popup=f"Date: {row['Date']}<br>Category: {row['Category']}",
        icon=folium.Icon(icon=icon, prefix='fa', color= color)
    ).add_to(m)


m
```
